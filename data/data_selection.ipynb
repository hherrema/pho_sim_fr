### Data Selection

Free recall studies to use in analysis of phonetic clustering.

- Intracranial
    - FR1
    - pyFR
    - IFR1
    
- Scalp
    - ltpFR2
        - sessions 0-22 (final sessions has different words presented)

```python
# imports
import numpy as np
import pandas as pd; pd.set_option('display.max_columns', None)
import cmlreaders as cml
from tqdm.notebook import tqdm
```

```python
df = cml.get_data_index()

df_intrac = df[df.experiment.isin(['FR1', 'pyFR', 'IFR1'])]
df_scalp = df[(df['experiment'] == 'ltpFR2') & (df['session'] != 23)]
```

```python
load_evs = 0
err_evs = 0
for _, row in tqdm(df_scalp.iterrows()):
    reader = cml.CMLReader(row.subject, row.experiment, row.session)
    
    try:
        evs = reader.load('events')
        load_evs += 1
    except BaseException as e:
        err_evs += 1
        
load_evs, err_evs
```
