### Data Selection

Free recall studies to use in analysis of phonetic clustering.

- Scalp: ltpFR2 (sessions 0-22, final session has different words presented)
        
- Intracranial: FR1, pyFR, IFR1

```python
# imports
import numpy as np
import pandas as pd; pd.set_option('display.max_columns', None)
import cmlreaders as cml
from tqdm.notebook import tqdm
```

```python
df = cml.get_data_index()

df_intrac = df[df.experiment.isin(['FR1', 'pyFR', 'IFR1'])]
df_scalp = df[(df['experiment'] == 'ltpFR2') & (df['session'] != 23)]
```

```python
df
```

```python
df_sel = cml.get_data_index('ltp')
```

```python
df_sel.groupby('experiment').agg(subject=('subject', 'nunique'), session=('session', 'count')).reset_index()
```

```python
df_select = df_sel.query("experiment == 'prelim'")
```

```python
df_scalp.subject.unique()
```

```python
df_select.subject.unique()
```

```python
all_words = []
for _, row in tqdm(df_select.iterrows()):
    reader = cml.CMLReader(row.subject, row.experiment, row.session)
    
    try:
        evs = reader.load('events')
        words = evs[(evs['type'] == 'WORD') | (evs['type'] == 'REC_WORD')].item_name.unique()
        all_words.extend(list(words))
    except BaseException as e:
        print(row.subject, row.experiment, row.session)
```
