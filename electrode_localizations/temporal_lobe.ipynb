## Electrode Localizations

Locate channels in the temporal lobe.

For pyFR, use `Loc2` field to match on temporal lobe.

```python
# imports
import pandas as pd; pd.set_option('display.max_columns', None)
import numpy as np
import cmlreaders as cml
from tqdm.notebook import tqdm
import warnings; warnings.filterwarnings("ignore")
```

```python
df = cml.get_data_index()

df_select = df[df.experiment.isin(['FR1', 'pyFR', 'IFR1'])]
df_select
```

#### pyFR

`Loc2` = `Temporal Lobe`.

```python
pyFR_t_lobe = []
for _, row in tqdm(df_select.query("experiment == 'pyFR'").iterrows()):
    reader = cml.CMLReader(row.subject, row.experiment, row.session, row.localization, row.montage)
    
    try:
        pairs = reader.load('pairs')
        t_lobe = pairs.query("Loc2 == 'Temporal Lobe'")
        
        if len(t_lobe) > 0:
            pyFR_t_lobe.append((row.subject, row.experiment, row.session, row.localization, row.montage, len(t_lobe)))
    except BaseException:
        continue
        
# save as dataframe
pyFR_t_lobe = pd.DataFrame(pyFR_t_lobe, columns=['subject', 'experiment', 'session', 'localization', 'montage', 'n_chans'])
```

#### FR1 and IFR1

```python
ind_region = []
wb_region = []
for _, row in tqdm(df_select.query("experiment != 'pyFR'").iterrows()):
    reader = cml.CMLReader(row.subject, row.experiment, row.session, row.localization, row.montage)
    
    try:
        pairs = reader.load('pairs')
        
        if 'ind.region' in pairs.columns:
            ind_region.extend(pairs['ind.region'].unique())
        
        if 'wb.region' in pairs.columns:
            wb_region.extend(pairs['wb.region'].unique())
            
    except BaseException:
        print(row.subject, row.experiment, row.session, row.localization, row.montage)
        
```
